# 16- Splitting a string into multiple columns

```python
# import libraries 
import pandas as pd 

df = pd.DataFrame({'name': ['john ibrahim', 'laj harani'],
                   'location': ['pakistan, lahore', 'CA , muku']})
df

```

```python
df.name.str.split(' ').head()
```

```python
# addiing  two splits into new columns
df[["first name", "last name"]] = df.name.str.split(' ', expand=True)
```

```python
df
```

```python
# split into two columns
df.name.str.split(' ', expand=True)
```

```python

df[["city", "country"]] = df.location.str.split(', ', expand=True)
```

```python
df
```

```python
# refine data manipulation
df = df[['first name','last name','city','country']]
df
```

# 23- RESHAPE A DATAFRAME

```python
import pandas as pd

fasla = pd.DataFrame([['12345', 100]],
                      columns=['zip','factory'])
fasla.head()
```

```python
fasla.head().T
```

```python
fasla.melt(id_vars='zip',var_name='loc_type')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='zip', y='factory', data=fasla)
plt.show()
```
