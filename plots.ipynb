# PLOTING IN PYTHON


```python
#IMPORT LIBRARIES 
import pandas as pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dtaset
kashti = sns.load_dataset('titanic')
kashti
```

```python
kashti.info()
```

```python
kashti.isnull().sum()
```

```python
kashti.survived.value_counts().plot(kind="bar")
```

```python
sns.countplot(x="survived", hue="sex", data=kashti)
```

```python
# scatter plot
plt.scatter(data=kashti, x="age", y="fare")
```

```python
sns.scatterplot(data=kashti, x="age", y="fare", hue="pclass")
```

```python
sns.barplot(data=kashti, x="sex", y="fare", hue="pclass")
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist(bins=10, figsize=(9,7), grid=False)
```

```python
sns.boxplot(data=kashti, x="sex", y="fare", hue="embark_town")
```

```python
sns.boxenplot(data=kashti, x="age", y="fare", hue="pclass")
```

```python
#assignment :How to change the x and y axis titles?
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.FacetGrid(kashti, hue="survived", col="pclass", margin_titles=True,
                  palette={1:"seagreen", 0:"gray"})

g = g.map(plt.scatter, "fare", "age", edgecolor="w")
g.add_legend()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.FacetGrid(kashti, hue="survived", col="sex", margin_titles=True,
                  palette={1:"seagreen", 0:"gray"})

g = g.map(plt.scatter, "fare", "age", edgecolor="w")
g.add_legend()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.FacetGrid(kashti, hue="survived", col="who", margin_titles=True,
                  palette={1:"seagreen", 0:"gray"})

g = g.map(plt.scatter, "fare", "age", edgecolor="w")
g.add_legend()
```

```python
sns.stripplot(data=kashti, x="sex", y="fare", hue="pclass")
```

```python
# heat map

corr = kashti.corr(numeric_only=True)
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("Heat map of kashti")
```

```python
sns.violinplot(data=kashti, x="sex", y="fare", hue="pclass")
```
